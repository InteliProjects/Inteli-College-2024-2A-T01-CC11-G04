```python
csv_path = "/content/drive/MyDrive/Mo패dulos/Mo패dulo 11/data/data_2024-09-24_13-13-45.csv"
df = pd.read_csv(csv_path)
df = df.drop('processed_question', axis=1)
df.sample(3)
```

```python
new_intents_path = "/content/drive/MyDrive/Mo패dulos/Mo패dulo 11/data/new_intents.csv"
df_n = pd.read_csv(new_intents_path)
df_n = df_n.rename(columns={'pergunta': 'question', 'intencao': 'intent'})
df_n.sample(3)
```
